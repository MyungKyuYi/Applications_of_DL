```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df_data = pd.DataFrame(iris.data, columns = iris_features)
df_labels = pd.DataFrame(iris.target, columns = ['label'])
```

```python
data=pd.concat([df_data,df_labels],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
data['label']=label.fit_transform(data['label'])
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in data.columns:
    data[i]=le.fit_transform(data[i])
```

```python
data
```

```python
X=data.drop('petal width (cm)',axis=1)
X.head()
```

```python
X=X.drop('label',axis=1)
X.head()
```

```python
y = data['petal width (cm)']
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
# Import library for Linear Regression
from sklearn.linear_model import LinearRegression

# Create a Linear regressor
lr = LinearRegression()

# Train the model using the training sets 
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test) 
y_pred
```

```python
lr.score(X_test, y_test)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Real Value")
plt.ylabel("Predicted Value")
plt.title("Real Value vs Predicted Value")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# 테스트 세트에 대한 예측을 만듭니다
y_prediction = lr.predict(X_test)

# 테스트 세트에 대한 평균 절댓값 오차를 계산합니다
mae = mean_absolute_error(y_test, y_prediction)
print(mae)
```

```python
print(lr.coef_, lr.intercept_)
```
